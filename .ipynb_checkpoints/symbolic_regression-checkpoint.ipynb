# [Genetic Programming with DEAP](https://deap.readthedocs.io/en/master/examples/gp_symbreg.html) 


```python
import operator
import math
import random

import numpy

from deap import algorithms
from deap import base
from deap import creator
from deap import tools
from deap import gp
```

## Creating the primitives set
Choosing the building blocks for the individual.

```python
def protected_div(left, right):
    try:
        return left / right
    except ZeroDivisionError:
        return 1

pset = gp.PrimitiveSet("MAIN", 1)
pset.addPrimitive(operator.add, 2)
pset.addPrimitive(operator.sub, 2)
pset.addPrimitive(operator.mul, 2)
pset.addPrimitive(protected_div, 2)
pset.addEphemeralConstant("eph_const", lambda: random.uniform(-1, 1))
pset.renameArguments(ARG0='x')
```

## [Creator](https://deap.readthedocs.io/en/master/api/creator.html#module-deap.creator)
The creator, as the name suggests, creates classes to use with evolutionary algorithms. In effect, new classes can be built from any imaginable type, from list to set, dict, PrimitiveTree and more, providing the possibility to implement genetic algorithms, genetic programming, evolution strategies, particle swarm optimizers, and many more.

```python
# Creates the fitness object (minimisation if the weight is negative)
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))

# Create the individual object
creator.create("Individual", gp.PrimitiveTree, fitness=creator.FitnessMin)
```

## Fitness Function

```python
def true_function(x):
    return x**4 - x**3 - x**2 - x

def evaluate_fitness(individual, X):
    # Compile the GP tree into a function
    func = toolbox.compile(expr=individual)
    
    sse = ((func(x) - true_function(x))**2 for x in X)
    
    # Fitness needs to be returned as an iterable according to DEAP doc.
    return math.fsum(sse) / len(X),
```

## Toolbox
In DEAP, parameters specific to the evolution process are registered through the toolbox. See [Operators](https://deap.readthedocs.io/en/master/api/tools.html#operators).

```python
X_train, X_val = [], []
for x in range(-10, 10):
    if random.random() < 0.3:
        X_val.append(x/10)
    else:
        X_train.append(x/10)

toolbox = base.Toolbox()
toolbox.register("expr", gp.genHalfAndHalf, pset=pset, min_=1, max_=2)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.expr)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("compile", gp.compile, pset=pset)
toolbox.register("evaluate", evaluate_fitness, X=X_train)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("mate", gp.cxOnePointLeafBiased, termpb=0.1)
toolbox.register("expr_mut", gp.genHalfAndHalf, min_=0, max_=2)
toolbox.register("mutate", gp.mutUniform, expr=toolbox.expr_mut, pset=pset)

# Decorate the mate and mutate method to limit the height of generated individuals
# Recommended max depth is 17 (see Koza '92)
toolbox.decorate("mate", gp.staticLimit(key=operator.attrgetter("height"), max_value=10))
toolbox.decorate("mutate", gp.staticLimit(key=operator.attrgetter("height"), max_value=10))
```

## Statistics
DEAP offers a simple class which can handle most of the “boring work”. This is done via [MultiStatistics](https://deap.readthedocs.io/en/master/api/tools.html#deap.tools.MultiStatistics).

```python
stats_fit = tools.Statistics(lambda ind: ind.fitness.values)
stats_size = tools.Statistics(len)
mstats = tools.MultiStatistics(fitness=stats_fit, size=stats_size)
mstats.register("avg", numpy.mean)
mstats.register("std", numpy.std)
mstats.register("min", numpy.min)
mstats.register("max", numpy.max)
```

## The Evolution

```python
pop = toolbox.population(n=300) # 300 individuals

# The hall of fame is a specific structure which contains the n best individuals 
hof = tools.HallOfFame(1) 
pop, log = algorithms.eaSimple(pop, 
                               toolbox, 
                               0.8, 
                               0.1, 
                               1000, 
                               stats=mstats,
                               halloffame=hof, 
                               verbose=True)
```

## Printing the best tree

```python
str(hof[0])
```

## Plotting the best tree
The function [deap.gp.graph()](https://deap.readthedocs.io/en/master/api/gp.html#deap.gp.graph) returns the necessary elements to plot tree graphs using NetworX or pygraphviz. The graph function takes a valid PrimitiveTree object and returns a node list, an edge list and a dictionary associating a label to each node. It can be used like following with pygraphviz.

```python
import networkx as nx
import matplotlib.pyplot as plt
from deap.gp import graph

import os     
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'

nodes, edges, labels = graph(hof[0])

g = nx.Graph()
g.add_nodes_from(nodes)
g.add_edges_from(edges)
pos = nx.nx_pydot.graphviz_layout(g, prog="dot")

nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_labels(g, pos, labels)
plt.axis("off")
plt.show()
```

## Testing the best tree

```python
func = toolbox.compile(expr=hof[0])

fig, axes = plt.subplots(ncols=3, figsize=(12, 6))
axes[0].set_title('Training')
axes[0].scatter(X_train, [func(x) for x in X_train], label='Genetic Programming')
axes[0].scatter(X_train, [true_function(x) for x in X_train], label='x**4 + x**3 + x**2 + x')

axes[1].set_title('Validation')
axes[1].scatter(X_val, [func(x) for x in X_val], label='Genetic Programming')
axes[1].scatter(X_val, [true_function(x) for x in X_val], label='x**4 + x**3 + x**2 + x')

X_test = [x/10 for x in range(-100, 100)]
axes[2].set_title('Test')
axes[2].scatter(X_test, [func(x) for x in X_test], label='Genetic Programming')
axes[2].scatter(X_test, [true_function(x) for x in X_test], label='x**4 + x**3 + x**2 + x')

plt.legend(loc='best')
plt.show()
```
